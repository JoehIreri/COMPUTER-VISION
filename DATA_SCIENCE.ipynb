```python
#Python Program to calculate the area of a triangle
# Takes inputs from the users
a = float(input('enter first side'))
b = float(input('enter second side'))
c = float(input('enter third number'))

#Calculate th semi perimeter
s = (a + b + c) / 2

#Calculate the area
area = (s*(s-a)*(s-b)*(s-c)) ** 0.5
print('The area of the triangle is %0.3f' %area)
```

```python
import matplotlib.pyplot as plt

# Data
vehicles = ["bus", "car", "bicycle", "walking", "train", "motorcycle"]
passengers_2022 = [24, 14, 12, 25, 20, 20]
passengers_2023 = [15, 48, 3, 5, 37, 2]

# Create bars
bar_width = 0.35
index = range(len(vehicles))

plt.bar(index, passengers_2022, bar_width, label='2022', color='cyan')
plt.bar([i + bar_width for i in index], passengers_2023, bar_width, label='2023', color='magenta')

# Set labels and title
plt.xlabel('Vehicles')
plt.ylabel('No. of Passengers')
plt.title('Commuter Travel in Nairobi (Kenya)')

# Set x-ticks and ticks labels
plt.xticks([i + bar_width / 2 for i in index], vehicles, rotation=45)

# Set y-ticks and ticks labels
plt.yticks([0, 10, 20, 30, 40])

# Add legend
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
